##### Se calcula la diferencia entre percentiles 75% y 25% utilizando estadisticas_descriptivas.loc['75%'] y estadisticas_descriptivas.loc['25%'], respectivamente. Esto proporciona una medida de dispersión entre el primer y tercer cuartil.

##### Además, se calcula la diferencia entre los valores máximos y mínimos de cada característica/variable utilizando estadisticas_descriptivas.loc['max'] y estadisticas_descriptivas.loc['min']. Esto permite evaluar la amplitud total de los datos.

##### Finalmente, se imprimen las diferencias entre los percentiles y los percentiles extremos para tener una idea de las diferencias entre diferentes puntos de corte en los datos.

##### Al observar estas diferencias, con esto se puede identificar si existen variaciones significativas entre los diferentes percentiles y si los valores máximos y mínimos están muy alejados de los demás valores, lo cual podría indicar la presencia de valores atípicos o una distribución sesgada.

```python
estadisticas_descriptivas = db.describe()

# Calcular diferencia entre percentiles
percentiles = [0.1, 0.25, 0.5, 0.75, 0.9]
diferencia_percentiles = estadisticas_descriptivas.loc['75%'] - estadisticas_descriptivas.loc['25%']
diferencia_percentiles_extremos = estadisticas_descriptivas.loc['max'] - estadisticas_descriptivas.loc['min']

# Imprimir diferencias entre percentiles
print("\nDiferencia entre percentiles:\n")
print(diferencia_percentiles)
print("\nDiferencia entre percentiles extremos:\n")
print(diferencia_percentiles_extremos)
```

#### Se crea una variable donde se guardan los valores de la columna price. 
#### Revisamos sus estadísticas descriptivas principales.

```python
precio = db['price']
precio.describe()
```

#### Miramos la frecuencia de cada uno los precios de los autos

```python
precios_unicos = precio.value_counts()
print('\n Frecuencia de los Precios \n')
precios_unicos
```

### Importamos la libreria Pyplot de Matplotlib para poder visualizar los precios

```python
import matplotlib.pyplot as plt
```

#### Primero vemos la distribución del 'Precio' en un Histograma

```python
precio.hist()
plt.xlabel('Rango de Precios')
plt.ylabel('Frecuencia de Precios')
plt.title('Distribución de los Precios en USD')
plt.show()
```

#### Ahora vemos los 'Precios' con un  gráfico Boxplot para ver los Outliers

```python
db.boxplot(column='price')
plt.xlabel('Conteo')
plt.ylabel('Precios')
plt.title('Distribución de los Precios')
plt.show()
```
